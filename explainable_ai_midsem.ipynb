Categorising some of the quantitative variables

```python
# Remove duplicates if any (based on entire row)
df = df.drop_duplicates()


# Create a new feature for glucose level category
def categorize_glucose(glucose):
    if glucose < 4:
        return 'Low'
    elif 4 <= glucose <= 7:
        return 'Normal'
    else:
        return 'High'

df['Glucose_Category'] = df['Glucose_Level'].apply(categorize_glucose)


# Columns like 'Patient_Number' and 'sample_id' might be redundant for analysis
columns_to_drop = ['Patient_Number', 'sample_id']

if 'Patient_Number' in df.columns:
  df.drop(columns='Patient_Number', inplace = True)
elif 'sample_id' in df.columns:
  df.drop(columns='sample_id', inplace = True)


# Average Blood Pressure (if 'SBP' and 'DBP' exist)
if 'SBP' in df.columns and 'DBP' in df.columns:
    df['Avg_Blood_Pressure'] = (df['SBP'] + df['DBP']) / 2

# Age Group Categorisation
def categorize_age(age):
    if age < 18:
        return 'Child'
    elif 18 <= age < 65:
        return 'Adult'
    else:
        return 'Senior'

df['Age_Group'] = df['Age'].apply(categorize_age)

# Drop irrelevant columns ( raw genotype count data if not needed for immediate analysis)
columns_to_drop = ['APOE2', 'APOE4']
df.drop(columns=columns_to_drop, inplace = True)


df.head()
```

Developing a KMeans model

```python
from sklearn.cluster import KMeans
```

Trying out the elbow method

```python
from sklearn.model_selection import train_test_split
```

```python
df
```

```python
X, xtest = train_test_split(df, test_size=0.05, random_state=42)
```

```python
import matplotlib.pyplot as plt
```

```python
num_samples = 10

# Range of K values to test (1 to num_samples)
K_range = range(1, num_samples + 1)

# List to store SSE for each K
sse = []

# Compute K-means clustering for each K and calculate SSE
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, init='k-means++', max_iter=100, n_init=1)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)  # inertia_ is the SSE

# Plot SSE vs. K
plt.figure(figsize=(8, 5))
plt.plot(K_range, sse, marker='o')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Sum of squared errors (SSE)')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
pip install kneed
```

```python
from kneed import KneeLocator

kneedle = KneeLocator(K_range, sse, S=1.0, curve="convex", direction="decreasing")

print(f'The optimal number of clusters is: {kneedle.elbow}')
```

```python
# On observing the graph, it seems that the optimal number of clusters is k = 2
```

Building the KMeans model with the optimal number of clusters

```python
# Number of clusters (K) for K-means
optimal_k = kneedle.elbow

# Train K-means model
model = KMeans(n_clusters=optimal_k, random_state=42, init='k-means++', max_iter=100, n_init=10)
model.fit(X)
```

```python
labels = model.labels_
```

KMeans Clustering Evaluation

```python
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import adjusted_rand_score
```

```python
silhouette = silhouette_score(X, labels)
calinski =  calinski_harabasz_score(X, labels)
wcss = model.inertia_
```

```python
print(f"Silhouette Score: {silhouette}")
print(f"Calinski-Harabasz Score: {calinski}")
print(f"With Cluster Sum of Scores (WCSS): {wcss}")
```
